## Problem 1: Counting DNA Nucleotides ##

```python
# First solution, O(n): Create initial counts, loop through dna string and add to counts respectively
a = c = g = t = 0
for base in dna:
    if base == 'A':
        a += 1
    if base == 'C':
        c += 1
    if base == 'G':
        g += 1
    if base == 'T':
        t += 1

print("DNA String: ", dna)
print("Base Counts: ", a, c, g, t)

# Second solution, O(4n) -> O(n): Use count function on each base
print("Base Counts: ", dna.count('A'), dna.count('C'), dna.count('G'), dna.count('T'))

# Third solution, also O(4n) -> O(n): Create map of the counts and print them iteratively with *
print("Base Counts: ", *map(dna.count, 'ACGT'))
```

## Problem 2: Transcribing DNA into RNA ##

```python
# O(n), Use string.replace function to replace all instances of substring 'T' with 'U'
rna = dna.replace('T', 'U')
print("DNA String: ", dna)
print("RNA String: ", rna)
```

## Problem 3: Complementing a String of DNA ##

```python
# Create translation table for the complements
# This part is O(1) since it's just creating the dict, no reliance on the length of the string
complement = str.maketrans("ACGT", "TGCA")

# apply the translation to each character in the string, then reverse the string
# O(n) for applying the complement table, O(n) for reversing, O(n) + O(n) = 2O(n), constants are dropped so just O(n)
print(dna.translate(complement)[::-1])
```
